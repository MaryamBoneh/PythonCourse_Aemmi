```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 12))

R = np.zeros((600,1200), dtype=np.uint8)
G = np.zeros((600,1200), dtype=np.uint8)
B = np.zeros((600,1200), dtype=np.uint8)

# red square
R[200:290, 200:290] = 255
G[200:290, 200:290] = 50

# green square
G[200:290, 300:390] = 255
R[200:290, 300:390] = 50

# blue square
B[300:390, 200:290] = 255
G[300:390, 200:290] = 160

# yellow square
G[300:390, 300:390] = 200
R[300:390, 300:390] = 255

image_result = cv2.merge((R, G, B))

image_result = cv2.putText(image_result, 'Microsoft', (420, 340), cv2.FONT_HERSHEY_SIMPLEX, 
                   4, (255, 255, 255), 10, cv2.LINE_AA)

for i in range(image_result.shape[0]):
    for j in range(image_result.shape[1]):
        if not image_result[i,j].any():
            image_result[i,j] = 70
            
    
plt.imshow(image_result)
```
