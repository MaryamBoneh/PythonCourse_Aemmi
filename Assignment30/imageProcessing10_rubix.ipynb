```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("rubix.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
plt.figure(figsize=(8, 8))

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        #  red color
        if image[i, j][0] < 50 and image[i, j][1] > 225 and image[i, j][2] > 225:
            image[i, j] = [255, 0, 0]
        
        #  green color
        elif image[i, j][0] > 225 and image[i, j][1] < 50 and image[i, j][2] > 225:
            image[i, j] = [0, 255, 0]
        
        #  blue color
        elif image[i, j][0] > 225 and image[i, j][1] > 225 and image[i, j][2] < 50:
            image[i, j] = [0, 0, 255]
plt.imshow(image)
```
