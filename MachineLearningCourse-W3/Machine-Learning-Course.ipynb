```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))
print('fit: ',r)

speed = myfunc(10)
print('speed of a 10 years old car: ', speed)

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
df = pandas.read_csv("cars.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

predictedCO2_2300kg = regr.predict([[2300, 1300]])
predictedCO2_3300kg = regr.predict([[3300, 1300]])

print(regr.coef_)
print('2300kg car _ coefficient values : ', predictedCO2_2300kg)
print('3300kg car _ coefficient values : ', predictedCO2_3300kg)
```

```python
scale = StandardScaler()
df = pandas.read_csv("cars.csv")
X = df[['Weight', 'Volume']]
y = df['CO2']
scaledX = scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaledX, y)

scaled = scale.transform([[2300, 1.3]])
predictedCO2 = regr.predict([scaled[0]])

print('scaledX: ', scaledX)
print('predictedCO2: ', predictedCO2)
```

```python
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]
test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))
myline = numpy.linspace(0, 6, 100)
train_r2 = r2_score(train_y, mymodel(train_x))
test_r2 = r2_score(test_y, mymodel(test_x))

plt.scatter(train_x, train_y) # blue point
plt.scatter(test_x, test_y) # orange point
plt.plot(myline, mymodel(myline)) # line

print('the R2 score when using training data:', train_r2)
print('the R2 score when using testing data:', test_r2)
print(mymodel(5))
plt.show()
```
